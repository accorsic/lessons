10. In this exercise we will use a simple "group by" to check the properties of some of the categorical variables in our data

a) What is the average price of the houses grouped by bedrooms?

```python
%%sql
select bedrooms, round(avg(price),2) as average_price, count(*) as count_houses
from house_price_data
group by bedrooms
order by bedrooms desc;
```

b) What is the average sqft_living of the houses grouped by bedrooms? 

```python
%%sql
select bedrooms, round(avg(sqft_living),2) as average_sqft_living, count(*) as count_houses
from house_price_data
group by bedrooms
order by bedrooms desc;
```

c) What is the average price of the houses with a waterfront and without a waterfront? 

```python
%%sql
select waterfront, round(avg(price),2) as average_price, count(*) as count_houses
from house_price_data
group by waterfront;
```

d) Is there any correlation between the columns condition and grade?

```python
%%sql
select `condition`, avg(grade) as Average_Grade, count(*) as Number_of_Houses
from house_price_data
group by `condition`
order by `condition`
```

11. Find a house with the following criteria
       - Number of bedrooms - either 3 or 4
       - Bathrooms - more than One Floor
       - No waterfront
       - Condition should be 3 at least
       - Grade should be 5 at least
       - Price less than 300000

```python
%%sql
select id from house_price_data
where (bedrooms=3 OR bedrooms=4) and (floors>1) and (waterfront=0) and (`condition`>=3) and (grade>=5) and (price<300000)
```

12. List of properties whose prices are twice more than the average of all the properties in the database

```python
%%sql
select id 
from house_price_data
where price >2*(
    select avg(price) as average_price
    from house_price_data)
```

13. Create a view called Houses_with_higher_than_double_average_price of the same query.

```python
%%sql
create view Houses_with_higher_than_double_average_price as
select id 
from house_price_data
where price >2*(
    select avg(price) as average_price
    from house_price_data)
```

```python
%%sql
select * from Houses_with_higher_than_double_average_price
```

14. What is the difference in average prices of the properties with three and four bedrooms?

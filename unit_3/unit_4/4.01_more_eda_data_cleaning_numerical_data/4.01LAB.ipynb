### 7. Plot every continuous variable. Comment what you can see in the plots.

```python
numericals = customer_df.select_dtypes(np.number)
for col in numericals.columns:
    sns.displot(numericals[col])
    plt.show()
```

#### 8. Do the same with the categorical variables (be careful, you may need to change the plot type to one better suited for continuous data!). Comment what you can see in the plots. You should also delete the column customer_id before you can try to use a for loop on all the categorical columns. Discuss why is deleting the column customer_id required.
```python
categoricals = customer_df.select_dtypes(np.object)
categoricals = categoricals.drop(['customer'], axis=1)
for col in categoricals.columns:
    sns.barplot(x = categoricals[col].value_counts().index, y = categoricals[col].value_counts())
    plt.show()
```

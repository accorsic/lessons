```python
def log_transfom_clean(x):
    x = np.log(x)
    if np.isfinite(x):
        return x
    else:
        return 0
```

```python
temp = list(map(log_transfom_clean, numericals['Customer Lifetime Value']))
sns.distplot(temp)
plt.show()
```
